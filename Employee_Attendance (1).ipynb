```python
import pandas as pd
import numpy as np
```

```python
file_path = "Employee_Attendance_Dataset (1).xlsx"
df = pd.read_excel(r"C:\Users\prerana\Downloads\Employee_Attendance_Dataset.xlsx")

df
```

```python
df.info()
```

```python
df["Attendance_Percentage"] = (df["Days_Present"] / df["Total_Working_Days"]) * 100
df
```

```python
average_attendance = df["Attendance_Percentage"].mean()
max_attendance = df["Attendance_Percentage"].max()
min_attendance = df["Attendance_Percentage"].min()

print("Average Attendance Percentage:", round(average_attendance, 2))
print("Maximum Attendance Percentage:", round(max_attendance, 2))
print("Minimum Attendance Percentage:", round(min_attendance, 2))
```

```python
dept_attendance = df.groupby("Department")["Attendance_Percentage"].mean()
dept_attendance
```

```python
low_attendance = df[df["Attendance_Percentage"] < 75]
low_attendance
```

```python
output_file = "Employee_Attendance_Processed.xlsx"
df.to_excel(output_file, index=False)

print("Processed file saved successfully.")
```
